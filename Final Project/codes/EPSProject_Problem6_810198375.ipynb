# Read CSV

```r
library(tidyverse)
  ## Read the csv file
setwd("C:/Users/Alireza/Documents")
df = read.csv("Automobile_data.csv")


df$bore<-as.numeric(as.character(df$bore))
df$stroke<-as.numeric(as.character(df$stroke))
df$horsepower<-as.numeric(as.character(df$horsepower))
df$peak.rpm<-as.numeric(as.character(df$peak.rpm))
df$price<-as.numeric(as.character(df$price))
df$normalized.losses<-as.numeric(as.character(df$normalized.losses))

df[complete.cases(df), ]
```

# Scatter Plotting for data

```r
ggplot(data = df) + geom_point(mapping = aes(x = length , y = width), color = "blue")
ggplot(data = df,
       aes(x = length,
           y =  width,
           color = num.of.doors)) +
  geom_point()
ggplot(data = df,
       aes(x = length,
           y =  width,
           color = body.style)) +
  geom_point()
ggplot(data = df,
       aes(x = engine.size,
           y =  width,
           color = engine.type)) +
  geom_point()
```

# More Plotting

```r
ggplot(data = df,
       aes(x = wheel.base,
           y =  length,
           color = make)) +
  geom_point()

ggplot(data = df,
       aes(x = wheel.base,
           y =  engine.size,
           color = engine.type)) +
  geom_point()

ggplot(data = df,
       aes(x = city.mpg,
           y =  engine.size,
           color = engine.type)) +
  geom_point()
```

# Jitter Plot

```r
ggplot(data = df) + geom_point(mapping = aes(x = city.mpg , y = engine.size), color = "red",position = "jitter")
```

# Find Hidden Relationships with Plots

```r
df %>% group_by(make) %>% group_by(count = n()) %>%  
    ggplot(aes(x= reorder(make,count),y=count)) + geom_bar(stat = "identity") + coord_flip() +
    xlab("Car Make") + ylab("Count")
ggplot(data = df,
       aes(x = fuel.type,
           y =  price)) +
  geom_boxplot() + xlab("Fuel Type") + ylab("Price")

ggplot((data = df),aes(x= make,y=price,fill = make)) + geom_boxplot() + coord_flip() +  xlab("Car Make") + ylab("Price")

  ggplot(df, aes(x="", y=engine.type, fill=engine.type)) +
  geom_bar(width = 1, stat = "identity") +
  coord_polar("y", start=0) +
  theme_void() + labs(title="Pie Chart for Engine Type")

ggplot(df, aes(sample = curb.weight)) + 
 stat_qq() + stat_qq_line() + labs(title="Normal Q-Q Plot",
       x="Normal Theoretical Quantiles",
       y="Curb Weight Sample")
  
ggplot(df, aes(x=engine.type, y=price, fill=engine.type)) +
geom_violin() + 
  labs(title="Violin plot", 
       subtitle="Engine Type and Price",
       x="Engine Type",
       y="Price")
```

# Linear Regression

```r
b <- (sd(df$width)/sd(df$length)) * cor(df$length,df$width)
a <- mean(df$width) - b * mean(df$length)
x <- seq(140,220,1)
plot(df$length, df$width, main = "Length and Width Plot",
     xlab = "Length", ylab = "Width",col = "red")
lines(x, b*x + a,col = "blue")
```
